```python
import matplotlib.pyplot as plt
import pandas as pd


#Kopie des annotated datensatzes erstellen
ab_ag_annotated_copy = pd.read_csv('../data/ab_ag_annotated.tsv', sep='\t')
#anzahl der eintr√§ge vor der L√∂schung zur Kontrolle
count_beforedeletion = len(ab_ag_annotated_copy)
#l√∂schung der eintr√§ge mit identischer sequenz
ab_ag_uniquesequences = ab_ag_annotated_copy.drop_duplicates(subset=['pdb', 'CDR_H1','CDR_H2','CDR_H3', 'CDR_L1', 'CDR_L2', 'CDR_L3'] )


#histogramm mit antigen h√§ufigkeiten


#H√§ufigkeiten der antigen_name
antigen_counts = ab_ag_uniquesequences["antigen_name"].value_counts()

#nur die Top 10 anzeigen
top_n = 10
antigen_counts = antigen_counts.head(top_n)

plt.figure(figsize=(12, 6))  # Breite x H√∂he von barplot

#Barplot => x=Antigen-Namen, y=H√§ufigkeit
plt.bar(antigen_counts.index, antigen_counts.values)

plt.xlabel("Antigen Name")
plt.ylabel("Anzahl der Antik√∂rper (H√§ufigkeit)")
plt.title("H√§ufigkeit der Antigen-Namen in ab_ag_uniquesequences")
plt.xticks(rotation=45, ha='right')  #x-Achsen-Labels drehen f√ºr bessere Lesbarkeit

plt.tight_layout()  #Optimiert Abst√§nde
plt.show()
```
